Lets create a Data set for visualize

```python
x = np.linspace(0,5,11)
y= x **2
```

MatPlotLib can be plot for 2 ways like 


1] Funtional Method


2] Object Oriented method

```python
# (1) Functional Method - Lets start with line plot
plt.plot(x,y) #we need and x and y to plot the chart 
plt.xlabel('This is x label') #we can assign the name to Axes
plt.ylabel('This is my Y label')
plt.title('My First CHart')
```

```python
# (2) Object Oriented Method - Here we are self 

fig = plt.figure(figsize=(12,8),dpi=100)
axes = fig.add_axes([0.5,0.5,0.8,0.5])
axes.plot(x,y)
axes.set_xlabel('This is my X axes')
```

```python
#Subplot

fig = plt.figure()
axes1=fig.add_axes([0.2,0.2,0.4,0.4])
axes2= fig.add_axes([0.3,0.3,0.2,0.2])

axes1.plot(x,y)
axes2.plot(y,x)

```

```python
#Subplot

plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(y,x)
plt.tight_layout
```

```python
fig = plt.figure()
axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig = plt.figure(figsize=(8,8))
axes = fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y,label='Normal chart')
axes.plot(x**2,y,label='x double')
axes.legend()
fig.show()
```

```python
fig = plt.figure(figsize=(8,8))
axes = fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y,label='Normal chart',color='red',linewidth=5,alpha=0.5,linestyle='-')
axes.plot(x**2,y,label='x double')
axes.legend()
fig.show()
```

```python
import pandas as pd
import numpy as np
import datetime
from datetime import date,timedelta
import matplotlib.pyplot as plt
import cufflinks as cf
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot

init_notebook_mode(connected=True)
cf.go_offline()
import yfinance as yf
```

```python
today = date.today() #featching todyas date from datetime library
print(today)
d1 = today.strftime('%Y-%m-%d') # format of date range we had mentioned 
end_date=d1  #This is pur end date i mean today, Now we have to choose the start date that is historical
print('This is my end date',end_date)

d2 = date.today()-timedelta(days=720)
d2 = d2.strftime('%Y-%m-%d')
start_date=d2
print('this is my end date for data',start_date)

data = yf.download('AAPL',start = start_date, end = end_date, progress=False)
print("data has been succesfully downloaded ")


```

```python
#lets see the view of data , how it looks 
data.head(2)
```

```python
import seaborn as sns

fig = plt.figure(figsize=(12,6)) #using matplotlib we defined the size of chart

sns.lineplot(x=data.index,y='Close',data=data) #using seaborn the chart has been drown
```

```python
px.line(data,x=data.index,y='Close',title='Time Series Analysis(yahoo)')
```

```python
data.Close.iplot(kind='line') # the same above chart was prepared by iplot i.e is Plotly which is very beautiful along with hower values 
```
